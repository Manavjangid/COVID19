```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('state_level_daily.csv')
```

```python
data.head(5)
```

```python
june = data.loc[data['Date'].str.contains('Jun')]
```

```python
july = data.loc[data['Date'].str.contains('Jul')]
```

```python
june_july = pd.concat((june, july), axis = 0)
```

```python
active = pd.DataFrame(june_july['Confirmed'] - june_july['Recovered'] - june_july['Deceased'], columns = ['Active'])
```

```python
active
```

```python
june_july = pd.concat([june_july, active], axis = 1)
```

```python
june_july.head()
```

```python
up_data = june_july.loc[june_july['State'].str.contains('UP')]
```

```python
up_data
```

```python
up_data.to_csv('up_data.csv')
```
